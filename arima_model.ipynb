#### AR(p) модель

$y_t = \alpha + y_{t-1} \phi_1 + ... + y_{t-p} \phi_p $

$\alpha$ - гиперпараметр
$y_{t-1}$ - вектор признаков $Ly$
$\phi_1$ - вектор весов для каждого признака  

Тогда нам необходимо оптимизировать матрицу весов $W$ размером $mxn$, где:
1. m - кол-во признаков
2. n = p (период функции)

$$W = \begin{pmatrix}w_{1,1} & w_{1, 2} & w_{1,3} & \dots & w_{1, m} \\  \dots & \dots & \dots \end{pmatrix}$$
